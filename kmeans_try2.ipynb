```python
import random
import math
import json
import sys
import time
import numpy as np
from scipy.spatial import distance
```

```python
# load data
file_name = 'data0.txt'
# data(list of list): [[index, dimensions], [.., ..], ...]
print('loading file...')
data = []              
fh = open(file_name)
for line in fh:
    line = line.strip().split(',')
    temp = [int(line[0])]
    for feature in line[1:]:
        temp.append(float(feature))
    data.append(temp)
data = np.array(data)

#get sample from data
length = len(data)
sample_size = int(length * 0.01)
random_nums = set()
sample_data = []

for i in range(sample_size):
  random_index = random.randint(0, length - 1)
  while random_index in random_nums:
      random_index = random.randint(0, length - 1)
  random_nums.add(random_index)
  sample_data.append(data[random_index])

print(length)
```

```python
#initialize variables and data structures

np.shape(sample_data)
# let, 
k = 4
n,d  = np.shape (sample_data)

centroids = np.array([[0 for _ in range(d)] for _ in range(k)])
cluster_affiliations = np.array([[sample_data[i], None] for i in range(n)])
```

```python
# pick k random datapoints as centroids
def initialize_centroids_simple( cent, clustaff, k ):
  all_indexes = list(np.arange(0,n))
  indexes = random.sample (all_indexes, k) 
  print('\nThese', k, 'indexes will be initialized as the first centroids:', indexes)

  for i, index in enumerate(indexes):
    print('point number', index, 'will be assigned as centroid number', i)
    temp_point = sample_data[index]
    for j in range(d):
      centroids[i][j] = sample_data[i][j] 
  return centroids

centroids = initialize_centroids_simple (centroids, cluster_affiliations, k)
centroid_indexes = [ centroids[i][0] for i in range(k) ]
print('these indexes have been intially centroided: ', centroid_indexes )


# REAFFILIATE DATAPOINTS
for thispoint_index in range (len(sample_data)):
  thispoint = sample_data[thispoint_index]
  distances = []

  # find distances from thispoint to all centroids
  for centroid_index in centroid_indexes:
    distances.append ( distance.euclidean(thispoint, data[centroid_index]) )

  # find the smallest distance, and corresponding cluster number. 
  closest_distance = np.amin(distances)
  min_dist_index = distances.index (closest_distance)
  # print('thispoints closest centroid is centroid number ', min_dist_index)

  # thispoint will be affiliated with cluster number min_dist_index.
  cluster_affiliations[thispoint_index][1] = min_dist_index
print('all points have been reclustered.')

print(cluster_affiliations[10][1])
cluster_affiliations[102][1]
```

```python
# once everything else is done, also define euclidean distance function
import numpy.linalg as LA 
def successCondition( indata ):
  sum = 0
  for datapoint in indata:
    temp = []
    for centroid in centroids:
      temp.append((LA.norm(datapoint - centroid ))**2)
    sum += min(temp)
    J = sum/n
  print(J)
  if (J<10**(-2)):
    print('Success condition is True. Stop kmeaning.')
    return True
  else:
    print('Success condition is false. Keep on kmeaning.')
    return False


successCondition(sample_data)
```
